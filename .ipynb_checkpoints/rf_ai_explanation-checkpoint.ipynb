````python
X,y = shap.datasets.boston()

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.05, random_state=0)
````

````python
# Random Forest
rf = sklearn.ensemble.RandomForestRegressor(n_estimators=100)
rf.fit(X_train, y_train)
rf.score(X_test,y_test)
````

````python
# Tree on Random Forest
explainerRF = shap.TreeExplainer(rf)
shap_values_RF_test = explainerRF.shap_values(X_test)
shap_values_RF_train = explainerRF.shap_values(X_train)

# Random Forest
df_shap_RF_test = pd.DataFrame(shap_values_RF_test, columns=X_test.columns.values)
df_shap_RF_train = pd.DataFrame(shap_values_RF_train, columns=X_train.columns.values)
````

# Permutation test

## How does it work?
It randomly picks a feature and makes the value constant and rescores the model and measuring the gap in performance it, it's indicative how important that feature is to the dataset. 

_Note: Using a held-out set makes it possible to highlight which features contribute the most to the generalization power of the inspected model. Features that are ranked high in importance on training set but not on the test might indicate an overfitted model to that feature._

## Collinear features? 
There won't be clear drop in the model performance it would be minor which would falsely indicate that feature isn't that important when it actually is.

```
4.2.2. Strongly correlated features¶
When two features are correlated and one of the features is permuted, the model will still have access to the feature through its correlated feature. This will result in a lower importance for both features, where they might actually be important. One way to handle this is to cluster features that are correlated and only keep one feature from each cluster. This use case is explored in: Permutation Importance with Multicollinear or Correlated Features.

```

## How to simplify when too many features?
Use dendrogram and drop the one of the correlated features. **not an issue with deeplearning*




````python
from sklearn.inspection import permutation_importance

plt.rcParams['figure.figsize'] = (8,3)

result = permutation_importance(rf, X_test, y_test, n_repeats=10, n_jobs=-1)
sorted_idx = result.importances_mean.argsort()

fig, ax = plt.subplots()
ax.boxplot(result.importances[sorted_idx].T,
           vert=False, 
           labels=X_test.columns[sorted_idx])
ax.set_title("Permutation Importances (test set)")
fig.tight_layout()
plt.show()
````

- Notice the placement of PTRATIO

````python
result = permutation_importance(rf, X_train, y_train, n_repeats=10, n_jobs=-1)
sorted_idx = result.importances_mean.argsort()

fig, ax = plt.subplots()
ax.boxplot(result.importances[sorted_idx].T,
           vert=False, 
           labels=X_test.columns[sorted_idx])
ax.set_title("Permutation Importances (train set)")
fig.tight_layout()
plt.show()
````

# What is a pdp?

It holds all other variables "fixed": uses averages all of the features as values and varies the target variable. 


The rug also helps in showing the data density in a particular area

````python
from sklearn.inspection import plot_partial_dependence

plt.rcParams['figure.figsize'] = (8,5)


plot_partial_dependence(rf, X_train, ['RM','LSTAT', 'DIS'],
                        n_jobs=-1, grid_resolution=20)

fig = plt.gcf()
fig.subplots_adjust(wspace=0.4, hspace=0.3)
````
